```python
#1导包 导入数据集
from sklearn.datasets import load_boston
```

```python
#数据加载
data=load_boston()
data
```

```python
X=data['data']
print(X)
y=data['target']
print(y)
```

```python
print(y.shape)
y=y.reshape(-1,1)
print(y.shape)
y
```

```python
#数据规范化
from sklearn.preprocessing import MinMaxScaler
ss=MinMaxScaler()
X=ss.fit_transform(X)
X
```

```python
#数据集的切分
import torch
X=torch.from_numpy(X).type(torch.FloatTensor)
print(X)
y=torch.from_numpy(y).type(torch.FloatTensor)
print(y)
```

```python
#切分数据集
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(X,y,test_size=0.2)
```

```python
print(train_x.shape)
print(test_x.shape)
print(train_y.shape)
print(test_y.shape)
```

```python
from torch import nn
#构造网络
model=nn.Sequential(
                    nn.Linear(13,16),
                    nn.ReLU(),
                    nn.Linear(16,1)
                )
```

```python
#构造优化器和损失函数
criterion=nn.MSELoss()
optimizer=torch.optim.Adam(model.parameters(),lr=0.01)
```

```python
#训练
max_epoch=300
iter_loss=[]
for i in range(max_epoch):
    #前向传播
    y_pred=model(train_x)
    #计算loss
    loss=criterion(y_pred,train_y)
    print("第{}次迭代的loss是:{}".format(i,loss))
    iter_loss.append(loss.item())
    #清空之前的梯度
    optimizer.zero_grad()    
    #反向传播
    loss.backward()
    #权重调整
    optimizer.step()
    
```

```python
#测试
output=model(test_x)
predict_list=output.detach().numpy()
print(predict_list)
```

```python
#绘制不同的iteration的loss
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(max_epoch)
y=np.array(iter_loss)
plt.figure()
plt.plot(x,y)
plt.title("在不同的迭代次数下的loss")
plt.xlabel("迭代次数")
plt.ylabel("loss")
plt.show()
```

```python
#查看真实值与预测值的散点图
x=np.arange(test_x.shape[0])
y1=np.array(predict_list)#预测值
y2=np.array(test_y)#实际值
line1=plt.scatter(x,y1,c="black")
line2=plt.scatter(x,y2,c='red')
plt.legend([line1,line2],["预测值","实际值"])
plt.title("预测值与实际值之间的误差")
plt.ylabel("波士顿房价")
plt.show()
```
