```python
import pandas as pd
pd.set_option('display.max_columns', 40)
#导入数据
train_data=pd.read_csv('used_car_train_20200313.csv',sep=' ')
train_data#训练集
```

```python
#同样把 test_data 也读取进来也处理一下
test_data=pd.read_csv('used_car_testB_20200421.csv',sep=' ')
test_data
```

```python
print(train_data['notRepairedDamage'].value_counts())#发现有缺失值，需要补全，
#采用众数补全的方法
train_data['notRepairedDamage'].replace('-','0.0',inplace=True)
train_data['notRepairedDamage'].value_counts()#缺失值已经补完了
```

```python
#看看测试里面是不是也缺失了，要是缺失了，用同样的方法
print(test_data['notRepairedDamage'].value_counts())#发现有缺失值，需要补全，
#采用众数补全的方法,同样的方法  复制粘贴过来，把train 换成test
test_data['notRepairedDamage'].replace('-','0.0',inplace=True)
test_data['notRepairedDamage'].value_counts()#发现有缺失值，需要补全
```

```python
#power的范围是0-600，大于600显然是不合理的，把大于600的设置成600
train_data['power'][ train_data['power'] >600]=600
test_data['power'][ test_data['power'] >600]=600
train_data['power'].describe()
```

```python
test_data['power'].describe()
```

```python
#将数据调整为float64
train_data['notRepairedDamage']=train_data['notRepairedDamage'].astype('float64')
test_data['notRepairedDamage']=test_data['notRepairedDamage'].astype('float64')
#缺失值用中位数补全
train_data=train_data.fillna(train_data.median())
test_data=test_data.fillna(train_data.median())

#数据类型转换完之后再看看info()
train_data.info()
print("分界线","-"*90)
test_data.info()#可以看到数据已经转化为float类型了
```

```python
features=train_data.columns
features  #看一下有哪些列  然后SaleID和price这两个是要去掉的
```

```python
features=features.tolist()
features.remove('price')
features.remove('SaleID')
features#可以看到已经没有 price和SaleID了
```

```python
from sklearn.preprocessing import MinMaxScaler
#特征归一化
min_max_scaler=MinMaxScaler()
x=min_max_scaler.fit_transform(train_data[features].values)
x_test=min_max_scaler.transform(test_data[features].values )
```

```python
#得到y值
y=train_data['price'].values
#数据切分
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.2)
```

```python
from tensorflow import keras
#搭建神经网络模型来训练
model=keras.Sequential([
    keras.layers.Dense(300,activation='relu',input_shape=[len( features)]),
    keras.layers.Dense(300,activation='relu'),
    keras.layers.Dense(200,activation='relu'),
    keras.layers.Dense(1), 
])
```

```python
#模型训练
model.compile(loss='mean_absolute_error',optimizer='Adam')
model.fit(train_x,train_y,batch_size=1024,epochs=200)
```

```python
#模型评估
from sklearn.metrics import mean_absolute_error
print("训练集MAE：{}".format(mean_absolute_error(train_y,model.predict(train_x)) ))
print("测试集MAE：{}".format(mean_absolute_error(test_y,model.predict(test_x)) ))
```

```python
#模型预测
predict_y=model.predict(x_test)
```

```python
#输出结果
result=pd.DataFrame()
result['SaleID']=test_data['SaleID']
result['price']=predict_y
#把预测的小于11的值改为11
result['price'][result['price']<11]=11
result.to_csv('result_nn.csv',index=False)#输出成文件
```
