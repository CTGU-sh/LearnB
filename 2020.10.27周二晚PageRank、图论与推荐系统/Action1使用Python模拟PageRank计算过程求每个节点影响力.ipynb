```python
import numpy as np

#定义转移矩阵
a = np.array(
    [ [0, 1/2, 1, 0], 
      [1/3, 0, 0, 1/2],
      [1/3, 0, 0, 1/2],
      [1/3, 1/2, 0, 0] ]
)
print("a的维度:{}".format(a.shape))
```

```python
#定义权重
b = np.array([1/4, 1/4, 1/4, 1/4])
w = b
print("w的维度:{}".format(w.shape))
```

```python
def work(a, w):
  #定义迭代函数
    for i in range(100):
        w = np.dot(a, w)
        print("第{}次迭代结果的权重w={}".format(i+1,w))
```

```python
#把上面定义好的转移矩阵和权重喂入迭代函数里面去
work(a,w)
#根据运行的结果可以看到迭代到20多次的时候就已经稳定了
```

```python
def random_work(a, w, n):
    #定义随机浏览模型
    d = 0.85
    for i in range(100):
        w = (1-d)/n + d*np.dot(a, w)
        print("第{}次迭代结果的权重w={}".format(i+1,w))
random_work(a, w, 4)#把定义好的转移矩阵和权重喂入迭代函数里面去
#可以看到也是20次左右的时候稳定了，但是权重相对于上面的有了一些变化
```

```python
#定义等级泄露矩阵，这里有一个黑洞，只进不出的，A，它对别人的贡献度为0
a_leak = np.array(
        [[0, 0, 0, 1/2], 
        [0, 0, 0, 1/2],
        [0, 1, 0, 0],
        [0, 0, 1, 0]])
random_work(a_leak, w, 4)
```

```python
#定义沉默矩阵，就是一个雷锋，只出不进，这里也是A
a_sink = np.array(
    [[0,  0, 0, 0], 
    [1/2, 0, 0, 1],
    [0,   1, 1, 0],
    [1/2, 0, 0, 0]])
random_work(a_sink, w, 4)
```
