```python
import pandas as pd 
import os

#读取土耳其新闻数据
news_TuErQi=open('news_土耳其.txt').read()
print(type(news_TuErQi))
news_TuErQi
```

```python
#读取武汉新闻数据
news_WuHan=open('news_武汉.txt').read()
print(type(news_WuHan))
news_WuHan
```

```python
#读取高考新闻数据
news_GaoKao=open('news_高考.txt').read()
print(type(news_GaoKao))
news_GaoKao
```

```python
from textrank4zh import TextRank4Keyword

# 输出关键词，设置文本小写，窗口为2,土耳其关键词
tr4w = TextRank4Keyword()
tr4w.analyze(text=news_TuErQi, lower=True, window=3)
print('土耳其新闻关键词：')
for item in tr4w.get_keywords(20, word_min_len=2):
    print(item.word, item.weight)
```

```python
# 武汉新闻关键词
tr4w = TextRank4Keyword()
tr4w.analyze(text=news_WuHan, lower=True, window=3)
print('武汉新闻关键词：')
for item in tr4w.get_keywords(20, word_min_len=2):
    print(item.word, item.weight)
```

```python
# 高考新闻关键词
tr4w = TextRank4Keyword()
tr4w.analyze(text=news_GaoKao, lower=True, window=3)
print('高考新闻关键词：')
for item in tr4w.get_keywords(20, word_min_len=2):
    print(item.word, item.weight)
```
