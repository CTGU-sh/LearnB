```python
import pandas as pd
#读取文件
file='jobs_4k.xls'
content=pd.read_excel(file)
content
```

```python
#把技能和职位提取出来转成list
position_names=content['positionName'].tolist()
skill_lables=content['skillLables'].tolist()
print("职位：\n",position_names)
print("\n分割线","-"*130,"\n")
print("技能",skill_lables)
```

```python
from collections import defaultdict
skill_positon_graph=defaultdict(list)
for p,s in zip(position_names,skill_lables):
    skill_positon_graph[p]+=eval(s)
print(skill_positon_graph)    
```

```python
import networkx as nx
import matplotlib.pyplot as plt
G=nx.Graph(skill_positon_graph)

#matplotlib无法显示中文
plt.rcParams['font.sans-serif'] = ['KaiTi']
plt.rcParams['font.serif'] = ['KaiTi']
plt.rcParams['font.size'] =8
#这句话解决坐标轴负号乱码
plt.rcParams['axes.unicode_minus'] = False

#以选取的20个工作岗位为例
import random
sample_nodes=random.sample(position_names,k=5)
sample_nodes
```

```python
sample_nodes_connections=sample_nodes
#初始化节点（刚开始为随机的20个职位）
for p,skills in skill_positon_graph.items():
    if p in sample_nodes:
        sample_nodes_connections+=skills
#抽取G中的子图
sample_graph=G.subgraph(sample_nodes_connections)
plt.figure(figsize=(50,30))
pos=nx.spring_layout(sample_graph,k=1)
nx.draw(sample_graph,pos,with_labels=True,node_size=30,font_size=10)
plt.show()
```

```python
#使用PageRank方法,对核心能力和核心职位进行影响力排序
pr=nx.pagerank(G,alpha=0.9)
ranked_postion_and_ability=sorted([(name,value) for name,value in pr.items()],key=lambda x:x[1],reverse=True)
print(ranked_postion_and_ability)
```

```python
#特征x需要去掉salary字段
X_content=content.drop(['salary'],axis=1)
#目标
target=content['salary'].tolist()
#将X_content内容都拼接成字符串，设置为merged字段
X_content['merged']=X_content.apply(lambda x:''.join(str(x)),axis=1)
X_content['merged']
```

```python
X_content['merged'][0]
```

```python
X_string=X_content['merged'].tolist()
X_string
```

```python
import jieba
import re

#合并在一起
def get_one_row_job_string(x_string_row):
    job_string=' '
    for i ,element in enumerate(x_string_row.split('\n')):
        if len(element.split())==2:
            _,value=element.split()
            #i=0,为id字段，需要去掉
            if i==0:
                continue
            #只保存value
            job_string+=value
    return job_string
```

```python
def token(string):
    return re.findall('\w+',string)
```

```python
cutted_X=[]
for i,row in enumerate(X_string):
    job_string=get_one_row_job_string(row)
    cutted_X.append(' '.join(list(jieba.cut(''.join(token(job_string))))))
cutted_X
```

```python
#使用TFIDF
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer()
X=vectorizer.fit_transform(cutted_X)
X
```

```python
target[:10]
```

```python
import numpy as np
target_numical=[np.mean(list(map(float,re.findall('\d',s)))) for s in target]
target_numical
```

```python
Y=target_numical
#使用KNN模型
from sklearn.neighbors import KNeighborsRegressor
model=KNeighborsRegressor(n_neighbors=2)
model.fit(X,Y)
```

```python
def predict_by_label(test_string,model):
    #分词
    test_words=list(jieba.cut(test_string))
    #转换为tfidf向量
    test_vec=vectorizer.transform(test_words)
    #预测
    y_pred=model.predict(test_vec)
    return y_pred[0]
test='测试 北京 3年 专科'
print(test,predict_by_label(test,model))
```

```python
persons=['测试 北京 3年 专科',
'测试 北京 4年 专科',
'算法 北京 4年 本科',
'UI 北京 4年 本科',
"广州Java本科3年掌握大数据",
"沈阳Java硕士3年掌握大数据", 
"沈阳Java本科3年掌握大数据", 
"北京算法硕士3年掌握图像识别"]

for p in persons:
    print("{}的薪资预测为{}".format(p,predict_by_label(p,model)))
```
