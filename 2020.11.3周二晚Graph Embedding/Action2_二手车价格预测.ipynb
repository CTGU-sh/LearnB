```python
import pandas as pd
pd.set_option('display.max_columns', 40)
#导入数据
train_data=pd.read_csv('used_car_train_20200313.csv',sep=' ')
train_data#训练集
```

```python
#探索一下train_data数据
train_data['regDate']#时间都出来了，但是这些数字不能直接放到模型里面用，需要将其转换成数字，就是以最小的时间开始经历的时间
#所以找出这时间里面的 最小值，其他的值减去这个最小值，这一列的数据就处理完了
```

```python
import numpy as np 
np.min(train_data['regDate'])#找出这个最小的时间
```

```python
temp=pd.to_datetime(train_data['regDate'],format='%Y%m%d',errors='coerce')#转换为时间戳格式
temp
```

```python
min_date=pd.to_datetime('19910101',format='%Y%m%d')#将最小的时间转换为时间戳格式
min_date
```

```python
#相减一下看看
(temp-min_date).dt.days
```

```python
#然后将train_data['regDate']汽车注册时间 转换一下，减去最小的时间得到的结果,得到汽车的展示时间
train_data['regTime']=(pd.to_datetime(train_data['regDate'],format='%Y%m%d',errors='coerce')-min_date ).dt.days
train_data['regTime']
```

```python
#同理可以将train_data['creatDate']汽车售卖时间 转换一下，减去最小的时间得到的结果,得到汽车的损耗时间
train_data['creatTime']=(pd.to_datetime(train_data['creatDate'],format='%Y%m%d',errors='coerce')-min_date ).dt.days
train_data['creatTime']
```

```python
#creatTime-regTime就得到了汽车的使用时间
train_data['usedTime']=train_data['creatTime']-train_data['regTime']
train_data['usedTime']
```

```python
#同样把 test_data 也读取进来也处理一下
test_data=pd.read_csv('used_car_testB_20200421.csv',sep=' ')
test_data
```

```python
#同理测试集的数据也这样处理一下
test_data['regTime']=(pd.to_datetime(test_data['regDate'],format='%Y%m%d',errors='coerce')-min_date ).dt.days
test_data['creatTime']=(pd.to_datetime(test_data['creatDate'],format='%Y%m%d',errors='coerce')-min_date ).dt.days
test_data['usedTime']=test_data['creatTime']-test_data['regTime']
```

```python
#看一下那个 notRepairedDamage 的数据分布情况
train_data['notRepairedDamage'].value_counts()#发现有缺失值，需要补全，
```

```python
#采用众数补全的方法
train_data['notRepairedDamage'].replace('-','0.0',inplace=True)
train_data['notRepairedDamage'].value_counts()#缺失值已经补完了
```

```python
#看看测试里面是不是也缺失了，要是缺失了，用同样的方法
test_data['notRepairedDamage'].value_counts()#发现有缺失值，需要补全，
```

```python
#采用众数补全的方法,同样的方法  复制粘贴过来，把train 换成test
test_data['notRepairedDamage'].replace('-','0.0',inplace=True)
test_data['notRepairedDamage'].value_counts()#发现有缺失值，需要补全
```

```python
#将数据调整为float64
train_data['notRepairedDamage']=train_data['notRepairedDamage'].astype('float64')
test_data['notRepairedDamage']=test_data['notRepairedDamage'].astype('float64')
```

```python
#找到数值类型的特征
numerical_cols=train_data.select_dtypes(exclude='object').columns
numerical_cols
```

```python
#找到 object 类型的特征
train_data.select_dtypes(include='object').columns
#发现没有，因为全部已经转成数值类型了    你们训练集全是数值了   可以拿去训练了吧
```

```python
train_data['SaleID'].value_counts()#这个saleID 跟最后的预测price是没有关系的，就把他给去掉吧，还有那个price也给去掉吧
```

```python
#去掉SaleID和price 就得到了训练集
feature_cols=[col for col in numerical_cols if col not in ['SaleID','price'] ]
feature_cols
```

```python
#提取特征列
X_data=train_data[feature_cols]#训练集特征
Y_data=train_data['price']#训练集标签
X_test=test_data[feature_cols]#测试集特征
```

```python
#定义一个统计函数方便了解数据的分布
def show_stats(data):
    print("最小值：{}".format(np.min(data)))
    print("最大值：{}".format(np.max(data)))
    print("最大小值之差：{}".format(np.ptp(data)))
    print("均值：{}".format(np.mean(data)))
    print("标准差：{}".format(np.std(data)))
    print("方差：{}".format(np.var(data)))
```

```python
#训练集price的统计情况
show_stats(Y_data)
```

```python
#缺失值用-1补全
X_data=X_data.fillna(-1)
X_test=X_test.fillna(-1)
```

```python
#补全完了就可以用模型来预测了 先试试线性回归
from sklearn.linear_model import LinearRegression
model=LinearRegression()
print("生成的模型:{}".format(model))#模型已经造好了
model.fit(X_data,Y_data)#训练
pridict_y_xgb=model.predict(X_test)
print("xgboost的预测结果的统计情况:")
show_stats( pridict_y_xgb)
```
