```python
from datetime import datetime
import pandas as pd
#数据加载
def parse(x):
    return datetime.strptime(x,'%Y %m %d %H')
dataset = pd.read_csv('./raw.csv',  parse_dates = [['year', 'month', 'day', 'hour']], index_col=0, date_parser=parse)
dataset#这个是原始数据  ，NO这一列是没有意义的 ，把它drop掉
```

```python
dataset.drop('No', axis=1, inplace=True)#axis=1表示列 inplace=True表示替换到原来的
dataset
```

```python
# 列名替换
dataset.columns = ['pollution', 'dew', 'temp', 'press', 'wnd_dir', 'wnd_spd', 'snow', 'rain']
dataset.index.name = 'date'
dataset#这样就把上述的表给替换掉了名字了
```

```python
# 缺失值填充
dataset['pollution'].fillna(0, inplace=True)
dataset.isnull().any()#没有空值了
```

```python
# 去掉第一天数据（前24小时）
dataset = dataset[24:]#从24行到结束，也就是把01号的数据全部给去掉了
# 数据浏览
print(dataset.head(5))
dataset
```

```python
# 数据保存
dataset.to_csv('pollution.csv')
```
