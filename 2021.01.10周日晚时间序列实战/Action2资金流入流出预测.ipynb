```python
#数据加载
import pandas as pd

data=pd.read_csv('user_balance_table.csv')
data
```

```python
data=pd.read_csv('user_balance_table.csv',parse_dates=['report_date'])
data#report_date这一行是日期类型，这样设置一下就OK了，要不是把日期当成整形的数字了
```

```python
#只关注purchase_amt和redeem_amt
#在report_date中，整体的申购和赎回总量
total_balance=data.groupby(['report_date'])['total_purchase_amt','total_redeem_amt'].sum()
total_balance
```

```python
#提取purchase 和redeem
purchase=total_balance[['total_purchase_amt']]
redeem=total_balance[['total_redeem_amt']]
purchase
```

```python
import matplotlib.pyplot as plt
import statsmodels.api as sm 
#指定范围内的数据进行可视化
def plot_stl(data):
    #将数据分成3个component:trend(趋势),seasonal(季节) residual(残差)
    result=sm.tsa.seasonal_decompose(data,freq=30)    
    #可视化
    fig=plt.figure(figsize=(12,8))
    ax1=fig.add_subplot(311)
    ax2=fig.add_subplot(312)
    ax3=fig.add_subplot(313)
    
    result.trend.plot(ax=ax1,title='Trend')
    result.seasonal.plot(ax=ax2,title='Seasonal')
    result.resid.plot(ax=ax3,title='Residual')
plot_stl(purchase['total_purchase_amt'])
```

```python
#抽取2014-04-01到2014-04-30的数据
purchase2=purchase[(purchase.index >='2014-04-01')&(purchase.index <='2014-04-30')]
#画出4月份的走势
plt.figure(figsize=(20,6))
plt.plot(purchase2.total_purchase_amt)
date_range=pd.date_range('2014-04-01','2014-04-30')
plt.xticks(date_range,rotation=90)
plt.show()
```

```python
from statsmodels.tsa.arima_model import ARIMA
#选择合适的p q
import pmdarima as pm
pm.arima.ndiffs(purchase,test='adf')
```

```python
from itertools import product

ps=range(0,8)
qs=range(0,8)
parameters=product(ps,qs)
parameters_list=list(parameters)
print(parameters_list)
```

```python
results=[]
best_aic=float('inf')#正无穷
for param in parameters_list:
    try:
        model=ARIMA(purchase,order=(param[0],1,param[1])).fit()
    except ValueError:
        print('参数错误',param)
        continue
    aic=model.aic
    if aic<best_aic:
        best_model=model
        best_aic=aic
        best_param=param
#     results.append([param,model.aic])        
best_param
```

```python
#对购买进行预测，使用type'levels'进行反差分，还原为原始数据的预测值
purchase_pred=best_model.predict('2014-09-01','2014-09-30',typ='levels')
purchase_pred
```
